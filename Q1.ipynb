$\textbf{Q1}$

Numerically integrate the following using Midpoint, Trapezoidal and Simpson
techniques for $N = 8, 16$ and $24$ and compare the result (in tabular format) with
the actual analytical result.
$$\displaystyle{\int_{1}^4 \sqrt{1+\frac{1}{x}}dx}$$

```python
from My_Lib import*
import math

def f(x): #Defining the function
    return math.sqrt(1+1/x) 

N = [8, 16, 24] #values of N
Analytic_solution = 3.620184280 #Actual Value of Integral

#listing the erros and solutions
solution = []  
Modulus_of_Error=[] 

for i in N:
    sol_mid = integration_midpoint(f,4,1,i)
    solution.append(sol_mid)
    Modulus_of_Error.append(abs(Analytic_solution-sol_mid))
    

print("Actual analytical value of the integral", Analytic_solution )

print ("\n~~~~~~~~~~~~~~~~MIDPOINT METHOD~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("|| Value via Midpoint\t\t|| Modulus of Error\t||")
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
for i in range (0,len(N)):
    print("||\t%.9f"%solution[i],"(N =",N[i],")\t||\t","%.6f"%Modulus_of_Error[i],"\t||")
    
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")  


########################################Trapezoidal Rule######################################################

N = [8, 16, 24] #value of N
Analytic_solution = 3.620184280 # actual value

#listing the erros and solutions
solution = []  
Modulus_of_Error=[]  

for i in N:
    sol_trap = integration_trap(f,4,1,i)
    solution.append(sol_trap)
    Modulus_of_Error.append(abs(Analytic_solution-sol_trap))
    



print ("\n~~~~~~~~~~~~~~Trapezoidal METHOD~~~~~~~~~~~~~~~~~~~~~~~~~")
print("|| Value via Trapezoidal\t|| Modulus of Error\t||")
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
for i in range (0,len(N)):
    print("||\t%.9f"%solution[i],"(N =",N[i],")\t||\t","%.5f"%Modulus_of_Error[i],"\t||")
    
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

##########################################################################################################
#Simpson Method
N = [8, 16, 24]
Analytic_solution = 3.620184280

#listing the erros and solutions
solution = []  
Modulus_of_Error=[]

for i in N:
    sol_simp = integral_simp(f,4,1,i)
    solution.append(sol_simp)
    Modulus_of_Error.append(abs(Analytic_solution-sol_simp))



print ("\n~~~~~~~~~~~~~~~~Simpson METHOD~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("|| Value via Simpson   \t\t|| Modulus of Error\t||")
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
for i in range (0,len(N)):
    print("||\t%.9f"%solution[i],"(N =",N[i],")\t||\t","%.6f"%Modulus_of_Error[i],"\t||")
    
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```
