$\textbf{Q3}$

Use Monte Carlo to estimate the value of $\pi$ from the following integral:
$$\displaystyle{\int_{0}^{1} \frac{4}{1+x^2} dx=\pi}$$


Use system built random number generator and sample $N$ starting from $10$ and
keep increasing in multiple of $10$. Go as far as possible within a reasonable time.
Plot $\pi ~\text{vs}.~ N$.

```python
import matplotlib.pyplot as plt

from My_Lib import*

Range = 10**4

def f(x):
    return 4/(1+x**2)

list_1 = [ i for i in range(10,Range, 10) ]
list_2 = []
for n in list_1:
    answer ,p = monte_carlo(f,0,1,n)
    list_2.append(answer)

print("The Following table shows the convergence of the Integral values..for N=10^4")
print ("\n~~~~~~~~~~~~~~~~~~~~Monte CARLO~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
print("||\tN\t\t|| Value of Integral\t\t||")
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
for i in range (0,len(list_1)):
    print("||\t%.9f"%list_1[i],"\t||\t",list_2[i],"\t||")
    
print ("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")




plt.title("$\pi$ vs $N$")
plt.plot(list_1,list_2,'r-.')
plt.xlabel("N")
plt.ylabel("Value of the Integral$")
plt.show()
```
